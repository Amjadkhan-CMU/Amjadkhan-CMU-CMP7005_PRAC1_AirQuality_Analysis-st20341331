# 1. Data Selection and Merging

In this section, I combine the datasets from four selected monitoring stations in Beijing. Two stations (Nongzhanguan and Wanshouxigong) represent urban areas, while Shunyi and Dingling represent suburban areas.

The goal of this step is to import the datasets, label them properly, and merge them into one single structured dataset that will be used for further analysis and modelling.

```python
import pandas as pd
import numpy as np
```

```python
base_url = "https://raw.githubusercontent.com/Amjadkhan-CMU/Amjadkhan-CMU-CMP7005_PRAC1_AirQuality_Analysis-st20341331/main/data/raw/"
```

```python
nong = pd.read_csv(base_url + "PRSA_Nongzhanguan_2013_2017_Urban.csv")
wan = pd.read_csv(base_url + "PRSA_Wanshouxigong_2013_2017_Urban.csv")
shu = pd.read_csv(base_url + "PRSA_Shunyi_2013_2017_SubUrban.csv")
ding = pd.read_csv(base_url + "PRSA_Dingling_2013_2017_SubUrban.csv")

print("Nongzhanguan:", nong.shape)
print("Wanshouxigong:", wan.shape)
print("Shunyi:", shu.shape)
print("Dingling:", ding.shape)
```

```python
nong["station"] = "Nongzhanguan"
wan["station"] = "Wanshouxigong"
shu["station"] = "Shunyi"
ding["station"] = "Dingling"

combined_df = pd.concat([nong, wan, shu, ding], axis=0)

print("Combined shape:", combined_df.shape)
```

```python
combined_df["datetime"] = pd.to_datetime(
    combined_df[["year", "month", "day", "hour"]]
)

combined_df = combined_df.sort_values("datetime")

combined_df.head()
```

After merging the four datasets, I created a unified datetime column using the year, month, day and hour variables.

This ensures the dataset is chronologically ordered and ready for time-based analysis such as hourly patterns, seasonal trends and long-term pollution changes.

```python
combined_df.to_csv("combined_air_quality_urban_suburban.csv", index=False)
```
